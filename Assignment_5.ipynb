```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
```

```python
sns.lineplot(x=x, y=y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line Plot')
plt.show()
```

```python
sns.barplot(x=x, y=y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Bar Plot')
plt.show()
```

```python
sns.scatterplot(x=x, y=y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
```

```python
plt.plot(x, y)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
plt.bar(x, y)
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
# Sample data
sizes = [30, 20, 15, 35]
labels = ['Category A', 'Category B', 'Category C', 'Category D']

# Plotting the pie plot
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Plot')
plt.show()
```
